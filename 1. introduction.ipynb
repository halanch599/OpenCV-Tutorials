```python
import cv2
```

#### Reading Images in OpenCV

```python
img = cv2.imread('./data/lena.jpg')

# img = cv2.imread('./data/lena.jpg',cv2.IMREAD_GRAYSCALE)
# B G R
# Grayscale
gray = cv2.imread('./data/lena.jpg',0)

cv2.imshow('Image', img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
# RGB
rgb = cv2.cvtColor(gray, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(gray, cmap='gray')
plt.show()
```

#### Write image to disk

```python
cv2.imwrite('./data/output.jpg', img)
```

```python
img = cv2.imread('./data/output.jpg')
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.axis('off')
plt.imshow(rgb)
plt.show()
```
